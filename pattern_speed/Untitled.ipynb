```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib import rc
import matplotlib as mpl
rc('text', usetex=True)
mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}']
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
```

```python
def read_index(idx):
    star = np.load('star_m12i_id' + str(idx)+'.npy')
    gas = np.load('gas_m12i_id' + str(idx)+'.npy')
    dark = np.load('dark_m12i_id' + str(idx)+'.npy')
#     return np.concatenate((star))
    return gas
```

```python
indices = np.arange(590, 601)
times = [13.77655729,
         13.77877760,
         13.78099605,
         13.78321595,
         13.78543563,
         13.78765511,
         13.78987438,
         13.79209261,
         13.79430981,
         13.79652845,
         13.79874688]
times = np.array(times)
times -= times[0]
times *= 1000
print(times)
```

```python
Cinit1 = 4.25
Cinit2 = 6.1
slope = -0.21
dphidt = 0.025
R = np.linspace(7.2, 9.2, 300)
```

```python
fig, ax = plt.subplots(5,2, figsize=(8, 20))
for idx, t, this_ax in zip(indices, times, ax.flatten()):
    test = read_index(idx)
    this_ax.hist2d(test[:,0], test[:,2], bins=300);

    phi1 = slope*R + Cinit1 + dphidt * t
    phi2 = slope*R + Cinit2 + dphidt * t
    this_ax.plot(R, phi1, c='r', alpha=0.5);
    this_ax.plot(R, phi2, c='r', alpha=0.5);
    
    phi1 = slope*R + Cinit1 + 1.3*dphidt * t
    phi2 = slope*R + Cinit2 + 1.3*dphidt * t
    this_ax.plot(R, phi1, c='r', alpha=0.5, ls='dashed');
    this_ax.plot(R, phi2, c='r', alpha=0.5, ls='dashed');

    phi1 = slope*R + Cinit1 + 0.7*dphidt * t
    phi2 = slope*R + Cinit2 + 0.7*dphidt * t
    this_ax.plot(R, phi1, c='r', alpha=0.5, ls='dashed');
    this_ax.plot(R, phi2, c='r', alpha=0.5, ls='dashed');
```
